```python
import pandas as pd
df = pd.read_csv ("D:\\NLP_SentimentA_RTE_Data\\SentimentData.csv")
df.head(10) 
```

```python
#Exteract positive reviews into file
for i in range(len(df)) :
    if (df.iloc[i, 1]=="positive"):
        filename="posfile"+str(i)+".txt"
        doc = df.iloc[i, 0]
        with open('D:\\NLP_SentimentA_RTE_Data\\pos\\'+filename, 'w', encoding='utf-8') as f:
            f.write(doc)
            
            f.close

for i in range(len(df)) :
    if (df.iloc[i, 1]=="negative"):
        filename="posfile"+str(i)+".txt"
        doc = df.iloc[i, 0]
        with open('D:\\NLP_SentimentA_RTE_Data\\neg\\'+filename, 'w', encoding='utf-8') as f:
            f.write(doc)
            
            f.close
```

```python
#Creating and writing text file to folder
with open('D:\\NLP_SentimentA_RTE_Data\\neg\\new4.txt', 'w') as f:
    f.write('Create a new text file!')
    
    f.close


with open('D:\\NLP_SentimentA_RTE_Data\\pos\\new4.txt', 'w') as f:
    f.write('Create a new text file!')
    
    f.close
```

```python
files_pos = os.listdir('D:\\NLP_SentimentA_RTE_Data\\pos\\')
files_pos = [open('D:\\NLP_SentimentA_RTE_Data\\pos\\'+f, 'r', encoding='utf-8').read() for f in files_pos]
files_neg = os.listdir('D:\\NLP_SentimentA_RTE_Data\\neg\\')
files_neg = [open('D:\\NLP_SentimentA_RTE_Data\\neg\\'+f, 'r', encoding='utf-8').read() for f in files_neg]
```

```python
len(files_pos), len(files_neg)
```

```python
pos_text = ""
neg_text = ""
```

```python
for t in files_pos:
    pos_text +=t

for t in files_neg:
    neg_text +=t
```

```python
wordcloud=WordCloud().generate(pos_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.savefig('C:/Users/tunji/Desktop/Postdoc Research/Concept_2/Tutor_Support/Data/InwordC1.jpg')
#plt.imsave('C:/Users/tunji/Desktop/Postdoc Research/Concept_2/Tutor_Support/Data/wrdCloud1.jpg')
plt.show()
```
